使用 [Lena](http://optipng.sourceforge.net/pngtech/img/lena.png)

- 改變圖片的 color space (`HLS`, `LAB`) 並呈現

```python
import cv2
import matplotlib.pyplot as plt


img_path = os.path.join(PRJ_DIR, 'data/lena.png')

img = cv2.imread(img_path, cv2.IMREAD_COLOR)

img_hsl = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
img_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsl_rgb = cv2.cvtColor(img_hsl, cv2.COLOR_HLS2RGB)
img_lab_rgb = cv2.cvtColor(img_lab, cv2.COLOR_LAB2RGB)

for i, (image, title) in enumerate(zip([img, img_hsl_rgb, img_lab_rgb], ['Original', 'HSL', 'LAB'])):
    plt.subplot(1, 3, i+1)
    plt.title(title)
    plt.imshow(image)
plt.show()
```
